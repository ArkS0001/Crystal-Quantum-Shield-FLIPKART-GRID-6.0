```python
import oqs

# Select the Dilithium algorithm
dilithium_alg = "Dilithium2"

# Generate key pair
with oqs.Signature(dilithium_alg) as signer:
    public_key = signer.generate_keypair()
    private_key = signer.export_secret_key()

# Save the keys for future use
with open("public_key.bin", "wb") as f:
    f.write(public_key)

with open("private_key.bin", "wb") as f:
    f.write(private_key)

# Message to be signed
message = b"Hello, this is a test message."

# Sign the message using the private key
with oqs.Signature(dilithium_alg) as signer:
    signer.import_secret_key(private_key)
    signature = signer.sign(message)

# Save the signature
with open("signature.bin", "wb") as f:
    f.write(signature)

# Verify the signature using the public key
with oqs.Signature(dilithium_alg) as verifier:
    verifier.import_public_key(public_key)
    is_valid = verifier.verify(message, signature)

if is_valid:
    print("Signature is valid.")
else:
    print("Signature is invalid.")
```
